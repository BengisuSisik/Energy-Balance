```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
# set parameters
L = 0.005 #length of the thermoelectric leg
nx = 6 # number of locations on the rod
dy = L / (nx - 1)  # distance between two consecutive locations
dx = 0.04
x = numpy.linspace(0, L, nx)
I = 0.242
R1 = dy / (1.2*dx*dx)
R = R1*numpy.ones(nx)
k = 1.2 #W/m*K
alpha = -30*10**(-6) #Seebeck coefficient
q_hot = 1 #W
q_cold = 1 #W !!!!!!!!!!!!!!!!
```

```python
nx = 6
A = 100*numpy.ones(nx) 
T = 298.15*numpy.ones(nx) #K
C = 10000
#print(A)
#print(T)

i = 0
while C > 0.0023:
    A = T.copy()
#    print('A at the beginning ')
#    print(A)
    i = i + 1
    print(i)
    T[0] = T[1] + ( q_hot + 0.5*R[0]*(I**2) - alpha*298.15*I ) / (k*dx/dy)
    
    for n in range(1, nx-1):
        
        T[n] = (T[n-1] + T[n+1])/2 + (0.5*(I**2)*R[n]*dy)/(2*k*dx) 
        
    T[nx-1] = T[nx-2] + (0.5*(I**2)*R[n] - q_cold + alpha*298.15*I) / (k*dx/dy)
    
    
    
 #    T[nx-1] = (T[nx-2] + (I**2*R[nx-1]*dy*dy)/(k) - (q_cold*dy)/(k))  / (1 + (alpha*I*dy/k) ) 
    
    
    print(T)
    
    C0 = T[0] - A[0]
    C1 = T[1] - A[1]
    C2 = T[2] - A[2]
    C3 = T[3] - A[3]
    C4 = T[4] - A[4]
    C5 = T[5] - A[5]
    
    C = C0 + C1 + C2 + C3 + C4 + C5
    
    print(C)
    
    T_avg = (T[0] + T[1] + T[2] + T[3] + T[4] + T[5] )/ 6
    
    print(T_avg)
```

```python
pyplot.title('Original Case')
pyplot.xlabel('Rod Length Top to Bottom [m]')
pyplot.ylabel('Temperature [K]')
pyplot.grid()
pyplot.plot(x, T);
```
